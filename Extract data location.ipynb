```python
blight_violations = pd.read_csv('./data/_97bd1c1e5df9537bb13398c9898deed7_detroit-blight-violations.csv', low_memory=False)
blight_violations
```

```python
blight_violations['ViolationStreetAddress'] = blight_violations['ViolationAddress'].map(lambda x: x.split('\n')[0])

def ExtractPositionInfo(info, position):
    try:
        positions = info[1:-1].split(',')
        return float(positions[position])
    except:
        return float('NaN')

blight_violations['ViolationLatitude'] = blight_violations['ViolationAddress'].map(lambda x: ExtractPositionInfo(x.split('\n')[2], 0))
blight_violations['ViolationLongitude'] = blight_violations['ViolationAddress'].map(lambda x: ExtractPositionInfo(x.split('\n')[2], 1))
blight_violations
```

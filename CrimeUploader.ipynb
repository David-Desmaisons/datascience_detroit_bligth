```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
```

```python
Crime = pd.read_csv('./data/_dcebfb2135a2bf5a6392493bd61aba22_detroit-crime.csv', low_memory=False)
Crime
```

```python
crimes = {cr for cr in Crime['CATEGORY']}
crimes = sorted(crimes)
orederedcrimes = { val : idx for idx, val in enumerate(sorted(crimes))} 
crimes
```

```python
import sqlite3

query = "insert into CrimeType (Name) values ('{}')"
    
con = sqlite3.connect("./data/location.s3db")
cur = con.cursor()
    
for crime in crimes:
    cur.execute(query.format(crime))
con.commit()
con.close()
```

```python
import sqlite3
import dateutil.parser

def Insert(row, idx, cur):  
    name = 'Crime{}'.format(idx+1)
    cur.execute("select Id from LocationReferenced where OriginId='{}'".format(name))
    idx = cur.fetchone()
    if idx is None:
        return
    index = idx[0]
    date = dateutil.parser.parse(row['INCIDENTDATE'])
    indextype = orederedcrimes[row['CATEGORY']]+1
    info ={'index':index, 'date':date, 'indextype': indextype}
    cur.execute('insert into CrimeData (LocationReferencedId, CrimeTypeId, Date) values (:index,:indextype,:date)', info)

con = sqlite3.connect("./data/location.s3db")
cur = con.cursor()

for index, row in Crime.iterrows():
    Insert(row, index, cur)
con.commit()
con.close()
```

```python
from collections import Counter, defaultdict

catcount = defaultdict(int)
#catcount = Counter()
for v in blight_violations['ViolDescription']:
    catcount[v]+=1
catcount
```

```python
from functools import reduce

def filtercoll(leng):
    return [(key,value) for key, value in catcount.items() if value>leng]


def countbylen(leng):
    coll = filtercoll(leng)
    return len(coll), reduce(lambda x,y: x+y, [c[1] for c in coll], 0)

countbylen(100)
```

```python
orederedviolation = { val : idx for idx, val in enumerate(fd)} 
orederedviolation
```

```python
orederedviolation['Other']=58
```

```python
ap = "PAID IN FULL"
par = "PARTIAL PAYMENT MADE"
notap = "NO PAYMENT APPLIED"
not2 = "NO PAYMENT ON RECORD"
```

```python
import sqlite3
import dateutil.parser
from datetime import date
import dateutil.parser

def Insert(row, idx, cur):  
    name = 'Violation_{}'.format(idx+1)
    cur.execute("select Id from LocationReferenced where OriginId='{}'".format(name))
    idx = cur.fetchone()
    if idx is None:
        return
    index = idx[0]
    currdate = row['TicketIssuedDT']
    try:
        currdate = dateutil.parser.parse(currdate)
    except:
        origin = date(1900, 1, 1)
        currdate = date.fromordinal(origin.toordinal()+int(currdate[6:11]))
    desc = row['ViolDescription'].title()
    indextype = orederedviolation[desc]+1 if desc in orederedviolation else orederedviolation['Other']+1
    payment = row['PaymentStatus']
    if payment == ap:
        payment = 2
    elif payment == par:
        payment =1
    elif payment == notap or payment==not2:
        payment =0
    else:
        raise Exception(payment)
    info ={'index':index, 'date':currdate, 'indextype': indextype, 'payment': payment}
    cur.execute('insert into ViolationData (LocationReferencedId, ViolationTypeId, Date, payment) values (:index,:indextype,:date, :payment)', info)

con = sqlite3.connect("./data/location.s3db")
cur = con.cursor()

for index, row in blight_violations.iterrows():
    Insert(row, index, cur)
con.commit()
con.close()
```

```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import sqlite3


blight_violations = pd.read_csv('./data/blight_violations_digested.csv',  low_memory=False)
blight_violations
```

```python
conn = sqlite3.connect('./data/location.s3db')
c = conn.cursor()
VioliationCount = 0


def Insert(row):
    try:  
        global VioliationCount
        VioliationCount = VioliationCount + 1
        name = str('Violation_' + str(VioliationCount))
        c.execute("insert into LocationReferenced (Latitude, Longitude, OriginId, Address) values(?, ?, ?, ?)", (row['ViolationLatitude'], 
                                                                                                                   row['ViolationLongitude'], name ,
                                                                                                                 row['ViolationStreetAddress']))
    except Exception as Error:
        print(Error)
  
for index, row in blight_violations.iterrows():
    Insert(row)
    
conn.commit()
c.close()
conn.close()
```

```python
Detroit_311 = pd.read_csv('./data/_dcebfb2135a2bf5a6392493bd61aba22_detroit-311.csv')
Detroit_311
```

```python
conn = sqlite3.connect('./data/location.s3db')
c = conn.cursor()
Count311 = 0

def clean(adrr):
    return adrr.replace(' Detroit,','').replace(' DETROIT,','').replace(' Michigan,','').replace(' Michigan','').replace(' MI,','').replace(' MI ',' ').replace(' Detroit','').replace(' DETROIT','').replace(' USA','').strip()


def Insert(row):
    try:  
        global Count311
        Count311 = Count311 + 1
        name = str('Call311' + str(Count311))
        address = clean(row['address'])
        c.execute("insert into LocationReferenced (Latitude, Longitude, OriginId, Address, OriginSourceId) values(?, ?, ?, ?, 2)", (row['lat'], 
                                                                                                                   row['lng'], name ,
                                                                                                                 address))
    except Exception as Error:
        print(Error)
  
for index, row in Detroit_311.iterrows():
    Insert(row)
    
conn.commit()
c.close()
conn.close()
```

```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import sqlite3
import geohash
import json
import ast


def Geocode(lat,long):
    try:
        return geohash.encode(lat,long,32)
    except:
        return ''

conn = sqlite3.connect('./data/location.s3db')
c = conn.cursor()
CrimeCount = 0

query = "insert into DismantledLocationReferenced (Latitude, Longitude, OriginId, Address, OriginSourceId, Location, GeoHash,Latitude_Min, Latitude_Max, Longitude_Min, Longitude_Max) values(?, ?, ?, ?, 5, ?, ?, ?, ?, ?, ?)"


def Insert(row):
    global CrimeCount
    CrimeCount = CrimeCount + 1
    name = str('Demolition' + str(CrimeCount))
    binfgeo = row['BuildInfo']
    correct = ast.literal_eval(binfgeo)['feature']['geometry'] if binfgeo != '{}' else None
    latitude = correct['y'] if correct is not None else row['Latitude']
    longitude = correct['x'] if correct is not None else row['Longitude']
    hashv = Geocode(latitude, longitude)
    Location= "({}, {})".format(latitude, longitude)
    c.execute(query, (latitude, longitude , name ,row['SITE_ADDRESS'], Location, hashv, row['ymin'], row['ymax'], row['xmin'], row['xmax']))
        
Demolition_Permits = pd.read_csv('./data/Demolition_Permits_digested_withbb.csv')
for index, row in Demolition_Permits.iterrows():
    Insert(row)
    
conn.commit()
c.close()
conn.close()
```
